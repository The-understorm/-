# Объяснение вводных данных и их вывод

1. Age - возраст человека
2. Sex - пол человека, где 0 - женщина, 1 - мужчина
3. Cp - Chest pain - тип боли в груди
4. Trtbs - Resting blood preasure - артериальное давление в состоянии покоя (в мм рт. ст.)
5. Chol - Сholesterol - холестерол, измеренный в мг/дл
6. Fbs - 1, если уровень сахара в крови натощак > 120 мг/дл, иначе 0
7. Restecg - Resting electrocardiographic results - результаты электрокардиографии в состоянии покоя
8. Thalachh - Thalium Stress Test result - максимальная частота сердечных сокращений
9. Exng - Exercise induced angina - 1, если стенокардия была вызвана физической нагрузкой, инчае 0
10. Oldpeak - предыдущий пик
11. Slp - подъем
12. Caa- количество крупных сосудов, принимает значаения 0 - 3
13. Thall - общее количество сосудов
14. Output - Целевая переменная

# Построение гипотез

## Гипотеза №1:
У людей со стенокардией часто наблюдется схожий повышенный уровень холестерина.

```python
figure(figsize=(12, 6), dpi=80)
sns.histplot(data['chol'], kde =True, color = 'blue')
sns.histplot(data[data['output']==1]['chol'], kde =True, color='red')
plt.title('Analysis of the connection between the amount of peolpe with stenocardia and the level of cholestrol')
plt.xlabel('Cholestrol')
plt.ylabel('Count')
plt.show()
```

### Вывод №1:
В идеальном случае уровень общего холестерина должен быть ниже 200 мг/дл, однако. Видно, что - содержание холестерина в количестве от 200 до 300 мг / дл часто присуще людям со стенокрадией на данном наборе данных, следовательно, у людей со стенокардией зачастую уровень холестерина превышает норму.

Гипотеза №1 подтверждена.


## Гипотеза №2:
Стенокардия чаще возникает у людей, ведущих малоподвижный образ жизни.

```python
#рассмотрим соотношение людей со стенокардией после физической нагрузки и без нее
with_execrise = np.sum(data['exng'] == True)
without_execrise = np.sum(data['exng'] == False)

#запишем переменные, которые считают общее количество людей со стенокардией
value_list = [with_execrise, without_execrise]
labels = 'The amount of people with stenocardia with exercises', 'The amount of people with stenocardia without exercises'
colors = ['blue', 'red']
fig, ax = plt.pie(value_list, colors=colors, shadow=True)
plt.legend(fig, labels, loc="best")

plt.tight_layout()
```

### Вывод №2:
Количество людей со стенокардией, ранее не имевших физическую нагрузку, больше чем людей, ранее имевших физическую нагрузку

Гипотеза №2 подтверждена.

## Гипотеза №3:
У людей со стенокардией высокое артериальное давление.

```python
figure(figsize=(10, 6), dpi=80)
data['trtbps'].value_counts().plot(kind='bar')
plt.xlabel('Blood Pressure')
plt.ylabel('Count')
plt.title('Blood pressure Analysis')
plt.show()
```

### Вывод №3:
120 - самое часто встречающееся значение на данном наборе данных, однако достаточно распространены и другие как более высокие, так и более низкие показатели артериального давления. Значит, нельзя выявить четкую взаимосвязь между значением артериального давления и стенокардией.

Гипотеза №3 опровергнута
